##### Analysis of the influence of large market capitalization stocks on the IHSG (Composite Stock Price Index) prices.
---

**Background:**

IHSG adalah indeks acuan rata-rata tertimbang seluruh saham di bursa indonesia, tercata per 12/05/2024 sudah ada lebih dari 800 perusahaan tercatat di bursa Indonesia.

Rumus perhitungan IHSG dapat ditemukan dengan cara berikut:

$$
\begin{align*}
IHSG &= \cfrac{Nilai\ Pasar}{Nilai\ dasar}{\times\ 100}\\
\end{align*}
$$


Nilai pasar sama dengan nilai kapitalisasi pasar, yang dapat dihitung dengan:

$$
Nilai\ pasar = p_1 \times q_1 + p_2 \times q_2 + p_3 \times q_3 + ... + p_n \times q_n
$$

Dimana:

$$
p = Closing\ Price \\
q = Jumlah\ saham\ yang\ tercatat
$$


---

Maka besaran market cap suatu saham dapat mempengaruhi harga IHSG.

```python
diff_price = pd.DataFrame()

for i in data_bigcap_2023.columns:
    if i != 'Date':
        diff_price[i] = data_bigcap_2023[i].diff()


diff_price.dropna(inplace=True)
diff_price
```

```python
for col in diff_price.columns:
    sns.histplot(data=diff_price[col])
    plt.show()
```

```python
mean = round(diff_price['BBRI'].mean(),3)
var = round(diff_price['BBRI'].var(),3)

print('mean:',mean,'var:',var)
```

```python
diff_price['BYAN'].min()
```

```python
def pdf_normal(x, mu, var):
    return (1 / np.sqrt(2 * np.pi * var)) *\
        np.exp(-(1 / (2 * var)) * (x - mu) ** 2)




for col in diff_price.columns:
    if col != 'IHSG':
        if col == 'TLKM':
            mean = round(diff_price[col].mean(),3)
            var_ = round(diff_price[col].var(),3)
            x_axis = np.arange(diff_price[col].min(), diff_price[col].max(), 10)
            y_axis = pdf_normal(x_axis, mean, var_)
            plt.plot(x_axis, y_axis)
            #plt.ylim(0,0.6)
            plt.xlabel(f"Harga saham {col} tahun 2023")
            plt.title('PDF distribusi normal dengan rata-rata = '+str(mean)+' dan varians = '+str(var_)+'\n standar deviasi = ' + str(np.round(np.sqrt(var_),3)))
            plt.show()
        else:
            mean = round(diff_price[col].mean(),3)
            var_ = round(diff_price[col].var(),3)
            x_axis = np.arange(diff_price[col].min(), diff_price[col].max(), 25)
            y_axis = pdf_normal(x_axis, mean, var_)
            plt.plot(x_axis, y_axis)
            #plt.ylim(0,0.6)
            plt.xlabel(f"Harga saham {col} tahun 2023")
            plt.title('PDF distribusi normal dengan rata-rata = '+str(mean)+' dan varians = '+str(var_)+'\n standar deviasi = ' + str(np.round(np.sqrt(var_),3)))
            plt.show()
```

```python
def mean_(kode):
    return round(diff_price[kode].mean(),3)

def var_(kode):
    return round(diff_price[kode].var(),3)
```

```python
diff_price.columns
```

```python
# berapa peluang harga saham turun < 0 ticks
peluang_turun = {}
for col in diff_price.columns:
    x = 0
    std = np.sqrt(var_(col))
    X = (x-mean_(col))/std
    peluang_turun[col]= norm.cdf(X,mean_(col),std)

for x,y in peluang_turun.items():
    print(f'Peluang harga {x} turun adalah {round(y,2)} ')
```

```python
# berapa probabilitas saham naik lebih dari 100 ticks
p100={}
for col in diff_price.columns:
    x = 100
    mean = mean_(col)
    std = np.sqrt(var_(col))
    p100[col] = 1-norm.cdf(x,mean,std)

for x,y in p100.items():
    print(f'Peluang {x} naik lebih dari 100 ticks adalah {round(y,2)}')
```

```python
# berapa peluang harga naik antara 0 - 50 ticks

x1 = 0
x2 = 50
p = {}

for col in diff_price.columns:

    mean = mean_(col)
    std = np.sqrt(var_(col))
    p0 = norm.cdf(x1,mean,std)
    p150 = norm.cdf(x2,mean,std)
    p[col] = p150-p0


for x,y in p.items():
    print(f'Peluang {x} naik sebanyak 0-50 ticks adalah {round(y,2)}')
```

## Uji Hipotesis
---

$$H_0: \mu_{\ peningkatan\ Big\ cap} = \mu_{\ peningkatan\ IHSG}$$
$$H_1: \mu_{\ peningkatan\ Big\ cap} \not= \mu_{\ peningkatan\ IHSG}$$

```python
# Inisiasi niali alpha
alpha = 0.05
```

```python
# Melihat apakah memiliki variance yang berbeda antara X dan Y

y = diff_price['IHSG']
x = diff_price.iloc[:,:4]

for col in x.columns:
    print(f'Variance saham {col}: {np.var(x[col])}, Variance IHSG: {np.var(y)}')
```

Variance dari masing masing saham terhadap IHSG berbeda.

```python
for col in x.columns:
    stat, p = ttest_ind(a = x[col], 
                    b = y, 
                    equal_var=False, 
                    alternative='two-sided') # eaual_var= False karena varians kedua populasi berbeda

    print(f'Untuk saham {col}, nilai t-hitung {round(stat,2)}, nilai p-value {round(p,2)}')
    
    
    # Pengambilan Keputusan
    if p > alpha:
        print('Kurang cukup bukti untuk menolak H0')
        print('---'*10)
    else:
        print('Cukup bukti untuk menolak H0')
        print('---'*10)
```
